```python
def plt_feat1_feat2(feat1, feat2):
    # group the data(predictor, taget) variables for our feature of interest
    x_m = features[feat1].groupby(features['N23tmakeittotheUniversity']).apply(list)[1]
    y_m = features[feat2].groupby(features['N23tmakeittotheUniversity']).apply(list)[1]
    x_b = features[feat1].groupby(features['N23tmakeittotheUniversity']).apply(list)[0]
    y_b = features[feat2].groupby(features['N23tmakeittotheUniversity']).apply(list)[0]
    fig, ax = plt.subplots(figsize=(8,8))

    
    #create subplots showing the correlations between the 2 features with 2 labels
    plt.scatter(x_m, y_m, color='red', label='make it to the University')
    plt.scatter(x_b, y_b, color='blue',label='interacting with employers')
    plt.xlabel(feat1)
    plt.ylabel(feat2)
    plt.title("Correlation between {} vs. {}".format(feat1, feat2))
    plt.legend()
    plt.show()
    
plt_feat1_feat2("N18facetofaceCareerG", "N15CareerGuidancesessi")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
target = pd.DataFrame(features)
data = features.drop('N23tmakeittotheUniversity', axis=1)

#scale the input using StandardScaler()
X_standardized = StandardScaler().fit_transform(data.values)
pca = PCA(svd_solver='full')
X_pca = pca.fit_transform(X_standardized)
#merge the PCA-extracted features into a new dataframe
pca_df = pd.DataFrame(X_pca)
pca_df = pca_df.merge(target, left_index=True, right_index=True, how='left')
pca_df.head()
```

```python
X_pca = PCA().fit(X_standardized)
plt.plot(np.cumsum(X_pca.explained_variance_ratio_))
plt.xlabel("#Components")
plt.ylabel("Cumulative Explained Variance Ratio")
pca = PCA(n_components=6)
pca.fit(X_standardized)
transformed_df = pd.DataFrame(pca.transform(X_standardized), columns = ['PC1', 'PC2', 'PC3', 
                                       'PC4', 'PC5', 'PC6'])
transformed_df.head()
```
