```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


```

```python
datasets = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/Tonny School datasets/SecondarySchoolStudents.csv")
datasets.head().transpose()
```

```python
datasets = datasets[(datasets['N12CareerGuidsveryimport'] == '1-Strongly agree') | (datasets['N12CareerGuidsveryimport'] == '2-agree') | (datasets['N12CareerGuidsveryimport'] == '3-Uncertain/not applicable') | (datasets['N12CareerGuidsveryimport'] == '4-Disagree') | (datasets['N12CareerGuidsveryimport'] == '5-Strongly disagree')]
# datasets = datasets[(datasets['QUESTIONAIREID'] > 0)]
top = datasets['N12CareerGuidsveryimport'].value_counts()
plt.figure(figsize=(10,7))
sns.barplot(x=top.index,y=top.values)
plt.xticks(rotation=45)
plt.title('Career Guidance is very important',color = 'blue')
```
